```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
points = [
     [5.9, 3.2]
    ,[4.6, 2.9]
    ,[6.2, 2.8]
    ,[4.7, 3.2]
    ,[5.5, 4.2]
    ,[5.0, 3.0]
    ,[4.9, 3.1]
    ,[6.7, 3.1]
    ,[5.1, 3.8]
    ,[6.0, 3.0]
]
redCluster = [6.2,3.2]
greenCluster = [6.6, 3.7]
blueCluster = [6.5,3.0]
```

```python
def euclidianDistance(point1,point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)
```

```python
for point in points:
    plt.scatter(point[0],point[1],c='k' ,marker = '^')

plt.scatter(redCluster[0],redCluster[1],c='r' ,marker = 'o')
plt.scatter(greenCluster[0],greenCluster[1],c='g' ,marker = 'o')
plt.scatter(blueCluster[0],blueCluster[1],c='b' ,marker = 'o')
plt.show()
```

```python
#Scatter plot with 
def plotWithRightColors(redCluster,greenCluster,blueCluster,points):
    for point in points: 
        redDistance = euclidianDistance(redCluster,point)
        greenDistance = euclidianDistance(greenCluster,point)
        blueDistance = euclidianDistance(blueCluster,point)
        minDistance = np.amin([redDistance,greenDistance,blueDistance])
        if redDistance == minDistance:
            plt.scatter(point[0],point[1],c='r' ,marker = '^')
        elif greenDistance == minDistance:
            plt.scatter(point[0],point[1],c='g' ,marker = '^')
        else:
            plt.scatter(point[0],point[1],c='b' ,marker = '^')

    plt.scatter(redCluster[0],redCluster[1],c='r' ,marker = 'o')
    plt.scatter(greenCluster[0],greenCluster[1],c='g' ,marker = 'o')
    plt.scatter(blueCluster[0],blueCluster[1],c='b' ,marker = 'o')
    plt.show()
    
```

```python
plotWithRightColors(redCluster,greenCluster,blueCluster,points)
```

```python
def calculateClusterIteration(redCluster,greenCluster,blueCluster,points):
        redPoints,greenPoints,bluePoints = ([],[],[])
        for point in points: 
            redDistance = euclidianDistance(redCluster,point)
            greenDistance = euclidianDistance(greenCluster,point)
            blueDistance = euclidianDistance(blueCluster,point)
            minDistance = np.amin([redDistance,greenDistance,blueDistance])
            if redDistance == minDistance:
                redPoints.append(point)
            elif greenDistance == minDistance:
                greenPoints.append(point)
            else:
                bluePoints.append(point)
        newRedCluster = np.sum(np.array(redPoints),axis=0)/len(redPoints)
        newGreenCluster = np.sum(np.array(greenPoints),axis=0)/len(greenPoints)
        newBlueCluster = np.sum(np.array(bluePoints),axis=0)/len(bluePoints)
        return newRedCluster.tolist(),newGreenCluster.tolist(),newBlueCluster.tolist()
```

```python
# inital iteration
plotWithRightColors(redCluster,greenCluster,blueCluster,points)
redCluster,greenCluster,blueCluster
```

```python
# First iteration
newRedCluster,newGreenCluster,newBlueCluster = calculateClusterIteration(redCluster,greenCluster,blueCluster,points)
plotWithRightColors(newRedCluster,newGreenCluster,newBlueCluster,points)
newRedCluster,newGreenCluster,newBlueCluster 
```

```python
#Second iteration
newRedCluster,newGreenCluster,newBlueCluster = calculateClusterIteration(newRedCluster,newGreenCluster,newBlueCluster,points)
plotWithRightColors(newRedCluster,newGreenCluster,newBlueCluster,points)
newRedCluster,newGreenCluster,newBlueCluster
```

```python
#Third iteration
newRedCluster,newGreenCluster,newBlueCluster = calculateClusterIteration(newRedCluster,newGreenCluster,newBlueCluster,points)
plotWithRightColors(newRedCluster,newGreenCluster,newBlueCluster,points)
newRedCluster,newGreenCluster,newBlueCluster
```
